#CPU

```python
cpu = vc[['เดือน','รายชื่อ','CPU','เป้าหมาย/CPU']]
cpu
```

```python
cpu['คิดเป็นเปอร์เซนต์'] = cpu['CPU']/200*100
cpu
```

```python
jan = cpu[cpu['เดือน']=='มกราคม'][['เดือน','รายชื่อ','CPU','เป้าหมาย/CPU']]
jan
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=[10,10]
```

```python
# ข้อมูล
data1 = jan['CPU']
labale = jan['รายชื่อ']
target = 200
fig, ax = plt.subplots()

# สร้างแท่งกราฟ
plt.bar(labale,data1,ec="#EBDD2B", fc="#7fffd4")
plt.title('มกราคม')

# กำหนดเส้นเป้าหมายด้วยวงกลมและเส้นต่อเนื่อง
ax.axhline(y=target, color='r', linestyle='--')
ax.plot([0, len(data1)], [target, target], 'r--')

# กำหนดชื่อแกน
ax.set_xlabel('รายชื่อ')
ax.set_ylabel('CPU')

# แสดงเลขบนแท่งกราฟ
for i, v in enumerate(data1):
    ax.text(i - 0.15, v + 1, str(v))

plt.show()
```
