#CPU

```python
cpu = vc[['เดือน','รายชื่อ','CPU','เป้าหมาย/CPU']]
cpu
```

```python
cpu['คิดเป็นเปอร์เซนต์'] = cpu['CPU']/200*100
cpu
```

```python
janc = cpu[cpu['เดือน']=='มกราคม'][['เดือน','รายชื่อ','CPU','เป้าหมาย/CPU']]
janc
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=[10,10]
```

```python
# ข้อมูล
datac1 = janc['CPU']
labale = janc['รายชื่อ']
target = 200
fig, ax = plt.subplots()

# สร้างแท่งกราฟ
plt.bar(labale,datac1,ec="#EBDD2B", fc="#7fffd4")
plt.title('มกราคม')

# กำหนดเส้นเป้าหมายด้วยวงกลมและเส้นต่อเนื่อง
ax.axhline(y=target, color='r', linestyle='--')
ax.plot([0, len(datac1)], [target, target], 'r--')

# กำหนดชื่อแกน
ax.set_xlabel('รายชื่อ')
ax.set_ylabel('CPU')

# แสดงเลขบนแท่งกราฟ
for i, v in enumerate(datac1):
    ax.text(i - 0.15, v + 1, str(v))

plt.show()
```
